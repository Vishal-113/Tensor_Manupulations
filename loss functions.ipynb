```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# 1. Define true values and model predictions
y_true = tf.constant([1, 0, 0], dtype=tf.float32)  # True class probabilities for one-hot encoded labels
y_pred = tf.constant([0.7, 0.2, 0.1], dtype=tf.float32)  # Predicted probabilities

# 2. Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE)
mse_loss = tf.keras.losses.MeanSquaredError()
cce_loss = tf.keras.losses.CategoricalCrossentropy()

mse_value = mse_loss(y_true, y_pred).numpy()
cce_value = cce_loss(y_true, y_pred).numpy()

print(f"Original MSE Loss: {mse_value}")
print(f"Original Cross-Entropy Loss: {cce_value}")

# 3. Modify predictions slightly
y_pred_modified = tf.constant([0.6, 0.3, 0.1], dtype=tf.float32)  # Slightly modified predictions
mse_value_modified = mse_loss(y_true, y_pred_modified).numpy()
cce_value_modified = cce_loss(y_true, y_pred_modified).numpy()

print(f"Modified MSE Loss: {mse_value_modified}")
print(f"Modified Cross-Entropy Loss: {cce_value_modified}")

# 4. Plot loss function values
labels = ['Original', 'Modified']
mse_values = [mse_value, mse_value_modified]
cce_values = [cce_value, cce_value_modified]

x = np.arange(len(labels))  # Label locations
width = 0.35  # Bar width

fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, mse_values, width, label='MSE Loss', color='blue')
bar2 = ax.bar(x + width/2, cce_values, width, label='Cross-Entropy Loss', color='orange')

# Add labels, title, and legend
ax.set_xlabel('Prediction Type')
ax.set_ylabel('Loss Value')
ax.set_title('Comparison of MSE and Cross-Entropy Loss')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
